## Slide 4: Python Code Example
Below is a Python code example to plot a sine wave.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the parabola
x = np.linspace(-10, 10, 200)  # 200 evenly spaced values from -10 to 10
y = x**2  # Parabola equation: y = x^2

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X values')
plt.ylabel('Y values (X^2)')
plt.title('Parabola: y = x^2')

# Show the plot
plt.grid(True)  # Add grid for better visualization
plt.show()
```
